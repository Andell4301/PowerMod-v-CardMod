I often find people get confused about what constitutes `cardMod` and what constitutes `powerMod` in the more sketchy cases like Oniland CE's, Space Ishtar's buff and Append skills. This short essay aims to help clarify the distinction.<br><br>

<b>CATEGORY 1: POWERMOD BUFFS</b><br><br>
This buff type is handled by the `powerMod` term in the following line of the [formula](https://github.com/atlasacademy/fgo-docs/blob/master/deeper/battle/damage.md):

```c
max([1 + powerMod + selfDamageMod + (critDamageMod * isCrit) + (npDamageMod * isNP)], 0.001)
```

Four types of buffs fall under the classification of `powerMod`:<br><br>

<b>Buff 1: DAMAGE</b><br>
This appears as either <b>[upDamage](https://apps.atlasacademy.io/db/NA/buffs?type=upDamage)</b> or <b>[downDamage](https://apps.atlasacademy.io/db/NA/buffs?type=downDamage).</b><br>
This is the standard `powerMod` buff everyone is familiar with. It's usually referred to as "STR Up" or "STR Down" in game.<br>
<b>Examples In Game:</b><br>
<b>[Str. Up vs Female](https://apps.atlasacademy.io/db/NA/buff/193)</b> on <b>[Jack The Ripper's NP](https://apps.atlasacademy.io/db/NA/noble-phantasm/600502)</b><br>
<b>[Str. Up vs Human](https://apps.atlasacademy.io/db/NA/buff/132)</b> on <b>[Command Code: Mortal](https://apps.atlasacademy.io/db/NA/command-code/1)</b><br>
<b>[Strength Up for Player Cards](https://apps.atlasacademy.io/db/JP/buff/385)</b>, the damage buff built in to <b>[Musashi's Fifth Force](https://apps.atlasacademy.io/db/NA/skill/327550)</b><br>

<b>Buff 2: DAMAGE INDIVIDUALITY</b><br>
This appears as either <b>[upDamageIndividuality](https://apps.atlasacademy.io/db/JP/buffs?type=upDamageIndividuality)</b> or <b>[downDamageIndividuality](https://apps.atlasacademy.io/db/JP/buffs?type=downDamageIndividuality).</b><br>
They sometimes use this when targeting a specific buff or debuff with `powerMod` (e.g., Str Up v Curse). It's not in NA right now, but a few servants have it in JP.<br>
<b>Examples In Game:</b><br>
<b>[Str. Up vs Curse](https://apps.atlasacademy.io/db/JP/buff/4693)</b> on <b>[Mochizuki Chiyome's S1](https://apps.atlasacademy.io/db/JP/skill/422352)</b><br>
<b>[Str. Up vs Burn](https://apps.atlasacademy.io/db/JP/buff/4695)</b> on <b>[Honey Lake](https://apps.atlasacademy.io/db/JP/skill/992240)</b><br>

<b>Buff 3: DAMAGE INDIVIDUALITY ACTIVE ONLY</b><br>
This appears as either <b>[upDamageIndividualityActiveOnly](https://apps.atlasacademy.io/db/JP/buffs?type=upDamageIndividualityActiveonly)</b> or <b>[downDamageIndividualityActiveOnly](https://apps.atlasacademy.io/db/JP/buffs?type=downDamageIndividualityActiveonly).</b><br>
This is the same as the previous buff type, but only targets buffs or debuffs applied by active skills, not passive skills. Currently, NA uses this exclusively instead of the previous one.<br>
<b>Examples In Game:</b><br>
<b>[Str. Up vs Debuff](https://apps.atlasacademy.io/db/JP/buff/1639)</b> on <b>[Arjuna Alter's S1](https://apps.atlasacademy.io/db/JP/skill/602650)</b><br>

<b>Buff 4: DAMAGE EVENT POINT</b><br>
This appears as <b>[upDamageEventPoint](https://apps.atlasacademy.io/db/JP/buffs?type=upDamageEventPoint)</b><br>
This is `powerMod` based on the amount of event points you have.<br>
<b>Examples In Game:</b><br>
<b>[Buster Strength Up](https://apps.atlasacademy.io/db/NA/buff/1382)</b> on the <b>[OniLand CE](https://apps.atlasacademy.io/db/NA/skill/963077)</b><br>
<b>[Buster Strength Up](https://apps.atlasacademy.io/db/JP/buff/3506)</b> on the <b>[Grail Live CE](https://apps.atlasacademy.io/db/JP/skill/992645)</b><br><br>

<b>FOR CALCING: ALL `powerMod` buff types applied via `PowerMod / Power / PMod / P`</b>

<br>
<b>CATEGORY 2: CARDMOD BUFFS</b>

This buff type is handled by the `cardMod` term in the following line of the [formula](https://github.com/atlasacademy/fgo-docs/blob/master/deeper/battle/damage.md):

```c
* {firstCardBonus + [cardDamageValue * max(1 + cardMod, 0)]}
```

Often, when people use the word `cardMod`, they think of what is listed as <b>Buff 1</b> here: the ordinary color specific card performance boost. This is correct, but there are actually other types of `cardMod` buff as well, all of which are calculated together:<br>

<b>APPLIED ON PLAYER:</b><br>
<b>Buff 1: COMMAND ALL</b><br>
This appears as either <b>[upCommandAll](https://apps.atlasacademy.io/db/NA/buffs?type=upCommandall)</b> or <b>[downCommandAll](https://apps.atlasacademy.io/db/NA/buffs?type=downCommandall).</b><br>
This is the standard color buff everyone is familiar with. This buff impacts Attack, NP Gen, and Star Gen.<br>
<b>Examples In Game:</b><br>
<b>[Buster Up](https://apps.atlasacademy.io/db/NA/buff/102)</b> on <b>[Ereshkigal's S2](https://apps.atlasacademy.io/db/NA/skill/447551)</b><br>
<b>[Extra Attack Up](https://apps.atlasacademy.io/db/NA/buff/30001)</b> on servants' Append 3.<br>
<b>FOR CALCING: Applied via `CardMod / CM / M`, or color versions e.g. `BCardMod / BCM / B`.</b>

<b>Buff 2: COMMAND ATK</b><br>
This appears as either <b>[upCommandAtk](https://apps.atlasacademy.io/db/NA/buffs?type=upCommandatk)</b> or <b>[downCommandAtk](https://apps.atlasacademy.io/db/NA/buffs?type=downCommandatk).</b><br>
This is the same as Buff 1, but it only impacts ATK, not NP Gen or Star Gen.<br>
<b>Examples In Game:</b><br>
<b>[Extra Attack Str Up](https://apps.atlasacademy.io/db/NA/buff/2537)</b> on <b>[Space Ishtar's NP](https://apps.atlasacademy.io/db/NA/noble-phantasm/1100901)</b><br>
<b>FOR CALCING: Applied via `CPower / CP`, or color versions e.g. `BPower / BCP / BP`.</b>

<b>Buff 3: COMMAND NP</b><br>
This appears as either <b>[upCommandNp](https://apps.atlasacademy.io/db/NA/buffs?type=upCommandnp)</b> or <b>[downCommandNp](https://apps.atlasacademy.io/db/NA/buffs?type=downCommandnp).</b><br>
This is the same as Buff 1, but it only impacts NP Gen, not ATK or Star Gen.<br>
<b>Examples In Game:</b><br>
No uses.

<b>Buff 4: COMMAND STAR</b><br>
This appears as either <b>[upCommandStar](https://apps.atlasacademy.io/db/NA/buffs?type=upCommandstar)</b> or <b>[downCommandStar](https://apps.atlasacademy.io/db/NA/buffs?type=downCommandstar).</b><br>
This is the same as Buff 1, but it only impacts Star Gen, not ATK or NP Gen.<br>
<b>Examples In Game:</b><br>
No uses.

<b>APPLIED ON ENEMY:</b><br>
<b>Buff 1: DEFENCE COMMAND ALL</b><br>
This appears as either <b>[upDefenceCommandAll](https://apps.atlasacademy.io/db/NA/buffs?type=upDefencecommandall)</b> or <b>[downDefenceCommandAll](https://apps.atlasacademy.io/db/NA/buffs?type=downDefencecommandall).</b><br>
This is the typical card resistance up or down everyone is familiar with. It impacts ATK, NP Gen, and Star Gen.<br>
<b>Examples In Game:</b><br>
<b>[Buster Resist Down](https://apps.atlasacademy.io/db/NA/buff/506)</b> on <b>[Karna's NP](https://apps.atlasacademy.io/db/NA/noble-phantasm/300402)</b><br>
<b>FOR CALCING: Applied via `CardMod / CM / M`, or color versions e.g. `BCardMod / BCM / B`.</b>

<b>Buff 2: DEFENCE COMMAND DAMAGE</b><br>
This appears as either <b>[upDefenceCommandDamage](https://apps.atlasacademy.io/db/JP/buffs?type=upDefencecommanDamage)</b> or <b>[downDefenceCommandDamage](https://apps.atlasacademy.io/db/JP/buffs?type=downDefencecommanDamage).</b><br>
This is the same as above, except it only impacts damage, not NP Gen or Star Gen.<br>
<b>Examples In Game:</b><br>
<b>[Buster Resist Up](https://apps.atlasacademy.io/db/JP/buff/535)</b> on <b>[This Ootakemaru NeroFest CQ](https://apps.atlasacademy.io/db/JP/skill/960970)</b><br>
<b>FOR CALCING: Applied via `CPower / CP`, or color versions e.g. `BPower / BCP / BP`.</b>
